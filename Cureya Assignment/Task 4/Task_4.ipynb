```python
from sklearn.preprocessing import  LabelEncoder

Lab_Ecode = LabelEncoder()
```

```python
Df.City = Lab_Ecode.fit_transform(Df.City)
```

```python
X = Df.drop(columns="Price",axis=1)
Y = Df.Price
```

```python
# Feature_importances_
from sklearn.ensemble import ExtraTreesRegressor

Df_Etr = ExtraTreesRegressor()
```

```python
Df_Etr.fit(X,Y)
```

```python
Df_Etr.feature_importances_
```

```python
Feature_Imp = pd.Series(Df_Etr.feature_importances_, index=X.columns) 

Feature_Imp.nlargest(20).plot(kind="barh")
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.20,random_state=5)
```

```python
print("Shape X_Train -",X_train.shape)
print("Shape Y_Train -",Y_train.shape)
print("Shape X_Test -",X_test.shape)
print("Shape Y_Test -",Y_test.shape)
```

### Linear Regression

```python
# Linear Regression

from sklearn.linear_model import LinearRegression

Df_Lr = LinearRegression()
```

```python
Df_Lr.fit(X_train,Y_train)
```

```python
# Test
Y_Test_Pred = Df_Lr.predict(X_test)
```

```python
# Train
Y_Train_Pred = Df_Lr.predict(X_train)
```

```python
from sklearn import metrics
from sklearn.metrics import mean_squared_error,r2_score
```

```python
# test
print("MAE",metrics.mean_absolute_error(Y_test,Y_Test_Pred))
print("MSE",metrics.mean_squared_error(Y_test,Y_Test_Pred))
print("RMSE",np.sqrt(metrics.mean_squared_error(Y_test,Y_Test_Pred)))
print("R2",metrics.r2_score(Y_test,Y_Test_Pred))
```

```python
# train
print("MAE",metrics.mean_absolute_error(Y_train,Y_Train_Pred))
print("MSE",metrics.mean_squared_error(Y_train,Y_Train_Pred))
print("RMSE",np.sqrt(metrics.mean_squared_error(Y_train,Y_Train_Pred)))
print("R2",metrics.r2_score(Y_train,Y_Train_Pred))
```

### Decision Tree

```python
# Decision Tree
from sklearn.tree import DecisionTreeRegressor
```

```python
Df_Dt = DecisionTreeRegressor()
```

```python
Df_Dt.fit(X_train,Y_train)
```

```python
Y_Test_Pred = Df_Dt.predict(X_test)
```

```python
# Decision Tree
print("MAE",metrics.mean_absolute_error(Y_test,Y_Test_Pred))
print("MSE",metrics.mean_squared_error(Y_test,Y_Test_Pred))
print("RMSE",np.sqrt(metrics.mean_squared_error(Y_test,Y_Test_Pred)))
print("R2",metrics.r2_score(Y_test,Y_Test_Pred))
```

### Random Forest Regression

```python
# Random Forest Regression
from sklearn.ensemble import RandomForestRegressor
```

```python
# Creating Object
Df_Rf=RandomForestRegressor(n_estimators=100,random_state=42)
```

```python
# fit the model
Df_Rf.fit(X_train,Y_train)
```

```python
Y_Test_Pred = Df_Rf.predict(X_test)
```

```python
# Random Forest
print("MAE",metrics.mean_absolute_error(Y_test,Y_Test_Pred))
print("MSE",metrics.mean_squared_error(Y_test,Y_Test_Pred))
print("RMSE",np.sqrt(metrics.mean_squared_error(Y_test,Y_Test_Pred)))
print("R2",metrics.r2_score(Y_test,Y_Test_Pred))
```
