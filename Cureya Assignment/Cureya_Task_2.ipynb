```python
X = Df.drop(columns="Outcome", axis=1)
Y = Df.Outcome
```

```python
print("Shape of X -",X.shape)
print("Shape of Y -",Y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25,random_state=0)
```

```python
print("Shape of X_Train -",X_train.shape)
print("Shape of X_Test -",X_test.shape)
print("Shape of Y_Train -",Y_train.shape)
print("Shape of Y_Test -",Y_test.shape)
```

```python
#  LogisticRegression
from sklearn.linear_model import LogisticRegression

Df_Lr = LogisticRegression()
```

```python
Df_Lr.fit(X_train,Y_train)
```

```python
Y_test_Pred = Df_Lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
confusion_matrix(Y_test,Y_test_Pred)
```

```python
accuracy_score(Y_test,Y_test_Pred)
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier

Df_Dt = DecisionTreeClassifier()
```

```python
Df_Dt.fit(X_train,Y_train)
```

```python
Df_test_Pred = Df_Dt.predict(X_test)
```

```python
confusion_matrix(Y_test,Y_test_Pred)
```

```python
accuracy_score(Y_test,Y_test_Pred)
```

```python
# Rando forest 
from sklearn.ensemble import RandomForestClassifier

Df_Rf = RandomForestClassifier()
```

```python
Df_Rf.fit(X_train,Y_train)
```

```python
Y_test_Pred = Df_Rf.predict(X_test)
```

```python
confusion_matrix(Y_test,Y_test_Pred)
```

```python
accuracy_score(Y_test,Y_test_Pred)
```

```python
# K-Means
from sklearn.cluster import KMeans

Df_Km = KMeans(n_clusters=2)
```

```python
Df_Km.fit(X_train,Y_train)
```

```python
Df_Km.cluster_centers_
```

```python
Df_test_Pred = Df_Km.predict(X_test)
```

```python
confusion_matrix(Y_test,Y_test_Pred)
```

```python
accuracy_score(Y_test,Y_test_Pred)
```
