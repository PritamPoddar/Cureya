```python
Df.drop(["Age",'JobLevel'],axis=1,inplace=True)
```

```python
# Import Label Encoder 

from sklearn.preprocessing import LabelEncoder
Lab_Code = LabelEncoder()
```

```python
for column in Df.columns:
    if Df[column].dtype==np.number:
        continue
    else:
        Df[column]=LabelEncoder().fit_transform(Df[column])
```

```python
X = Df.drop(columns="Attrition",axis=1)
Y = Df.Attrition
```

```python
# train_test_split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25,random_state=0)
```

```python
# LogisticRegression
from sklearn.linear_model import LogisticRegression
```

```python
Df_Lr = LogisticRegression()
```

```python
Df_Lr.fit(X_train,Y_train)
```

```python
Y_test_Pred = Df_Lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
# Confusion_matrix
confusion_matrix(Y_test,Y_test_Pred)
```

```python
# Accuracy_score
accuracy_score(Y_test,Y_test_Pred)
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
```

```python
Df_Rfc =  RandomForestClassifier()
```

```python
Df_Rfc.fit(X_train,Y_train)
```

```python
Y_test_Pred = Df_Rfc.predict(X_test)
```

```python
# Confusion_matrix
confusion_matrix(Y_test,Y_test_Pred)
```

```python
# Accuracy_score
accuracy_score(Y_test,Y_test_Pred)
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier
```

```python
Df_Dtc = DecisionTreeClassifier()
```

```python
Df_Dtc.fit(X_train,Y_train)
```

```python
Y_test_Pred = Df_Dtc.predict(X_test)
```

```python
# Confusion_matrix
confusion_matrix(Y_test,Y_test_Pred)
```

```python
# Accuracy_score
accuracy_score(Y_test,Y_test_Pred)
```

```python
# Support Vector Classifier) 
from sklearn.svm import SVC
```

```python
Df_Svc = SVC()
```

```python
Df_Svc.fit(X_train,Y_train)
```

```python
Y_test_Pred = Df_Svc.predict(X_test)
```

```python
# Confusion_matrix
confusion_matrix(Y_test,Y_test_Pred)
```

```python
# Accuracy_score
accuracy_score(Y_test,Y_test_Pred)
```
